```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from  selenium import webdriver
from  selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import time
from datetime import datetime
import gzip
from io import BytesIO
import pandas as pd
import xml.etree.ElementTree as ET
```

```python

url_basic = 'https://url.publishedprices.co.il/file/d/'

chrome_options = webdriver.ChromeOptions()
chrome_options.add_experimental_option('detach',True)

driver = webdriver.Chrome(options=chrome_options)

driver.get('https://url.publishedprices.co.il/login')

driver.find_element(By.NAME, value='username').send_keys('freshmarket')

driver.find_element(By.ID, value='login-button').click()
time.sleep(10)
soup = BeautifulSoup(driver.page_source, "html.parser")

td_shop = soup.find_all(name="a", class_= "f")

for i,td in enumerate(td_shop):
    
    file = td.get("title")
    if file[:6] == 'Stores' and file[-3:]=='xml':
        file_url = url_basic+file
        print(file_url) 


        cookies = driver.get_cookies()

        session = requests.Session()
        for cookie in cookies:
            session.cookies.set(cookie['name'], cookie['value'])

        response = session.get(file_url)


        xml_content = response.content.decode('utf-16')
    
        root = ET.fromstring(xml_content)

        data = []
        for Store in root.findall('.//Stores/Store'):
            row = {child.tag: child.text for child in Store}
            data.append(row)
            df = pd.DataFrame(data)
 
        break     

driver.quit()
        
df
```
